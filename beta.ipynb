```python
# import quandl
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta
import matplotlib.pyplot as plt
# %load keys.py
plt.style.use('fivethirtyeight')
# quandl.ApiConfig.api_key = "FqjcQ6Wyxf36_2fPqqUg"
#note: quandl wiki dataset no longer updated as of ~April 2018 it seems
import yfinance as yf
```

```python
#Parameters
ticker = yf.Ticker('MSFT')
benchmark = yf.Ticker('SPY')
# today = date.today()
```

```python
ticker_data = ticker.history(period="max")
benchmark_data = benchmark.history(period="max")
ticker_data.reset_index(level=0, inplace=True)
benchmark_data.reset_index(level=0, inplace=True)
ticker_data
```

```python
t = ticker_data.loc[:, ['Date', 'Close']]
b = benchmark_data.loc[:, ['Date', 'Close']]
begin = max(t.loc[0, 'Date'], b.loc[0, 'Date'])
t = t[t['Date'] >= begin]
b = b[b['Date'] >= begin]
t
```

```python
t['Change'] = t['Close'].pct_change()
b['Change'] = b['Close'].pct_change()
t.reset_index(drop=True, inplace=True)
b.reset_index(drop=True, inplace=True)
t.drop([0], inplace=True)
b.drop([0], inplace=True)
t
```

```python
ticker_change = t['Change'].values
bench_change = b['Change'].values
```

```python
beta = np.cov(ticker_change, bench_change)[0][1]/np.var(bench_change)
beta
```
